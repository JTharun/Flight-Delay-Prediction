```python
import pandas as pd
```

```python
file1='/content/drive/MyDrive/RegressionFullDataSet.csv'
```

```python
df=pd.read_csv(file1)
```

```python
df
```

```python
df['ArrDelayMinutes'].max()
```

```python
temp=[]
t=-1
for i in df['ArrDelayMinutes']:
  t+=1
  if i>=15 and i<300 :
    temp.append(t)
```

```python
len(temp)
```

```python
dictt={}
a=0
for i in temp:
  a+=1
  dictt[a]=df.loc[i]
  
```

```python
df_under300=pd.DataFrame(dictt)
```

```python
df_under300=df_under300.transpose()
```
