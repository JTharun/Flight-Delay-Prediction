```python
import pandas as pd
import json
import os
from glob import glob
import math
```

```python
path00="/content/drive/MyDrive/Weather/weather"
first_flow=os.listdir(path00)
```

```python
len_flow=0
final_flow={}
for i in first_flow:
  third_flow=[]
  second_flow=os.listdir(path00 +'/'+ i)
  for j in second_flow:
    if '2016' in j or '2017' in j:
      third_flow.append(j)
  
    final_flow[i]=third_flow
  len_flow+=len(third_flow)
  
```

```python
dfp2=pd.DataFrame()
for key ,value in final_flow.items():
  #for k in range(0,len(value)):
  for l in value:
    file1=open(path00 + '/' + key +'/' +l)
    j_load1=json.load(file1)
   
    dr1=j_load1['data']
    dfp1=pd.DataFrame()
    date_list=[]
    air_list=[]
    n1=len(dr1['weather'])              #works
    #for i1 in range(0,n1):
     # date=dr1['weather'][i1]['date']
      #date_list.extend([date for i1 in range(24)])
    air=key
    for j1 in range(0,n1):
      date=dr1['weather'][j1]['date']
      
      date_list.extend(date for i in range(24))
      air_list.extend(air for i in range(24))
      for g1 in range(0,24):
        dff1=pd.DataFrame(dr1['weather'][j1]['hourly'][g1])
        
        dfp1=dfp1.append(dff1, ignore_index = True)
    #print(len(dfp1))
    dfp1['Date']=date_list
    dfp1['Airport']=air_list
    dfp2 = dfp2.append(dfp1, ignore_index = True)
    #dfp2['Date']=date_list


 
```

```python
df_final=dfp2[['windspeedKmph', 'winddirDegree', 'weatherCode', 'precipMM',
'visibility' ,'pressure' ,'cloudcover' ,'DewPointF',
'WindGustKmph' ,'tempF' ,'WindChillF', 'humidity',
'Date' ,'time' ,'Airport']]
df_final.reset_index(drop=True, inplace=True)
```

```python
patht="/content/drive/MyDrive/Flight Data/2016"
EXT = "*.csv"
csv_files = [file
                 for path1, subdir, files in os.walk(patht)
                 for file in glob(os.path.join(path1, EXT))]
```

```python
pathh = "/content/drive/MyDrive/Flight Data/2017"
EXT = "*.csv"
csv_files2 = [file
                 for path, subdir, files in os.walk(pathh)
                 for file in glob(os.path.join(path, EXT))]
#print(csv_files2)
```

```python
see1=pd.DataFrame()

for i in range(0,len(csv_files)):
    
    df0 = pd.read_csv(csv_files[i])
    df01=df0[['FlightDate', 'Quarter' ,'Year','Month',
'DayofMonth' ,'DepTime' ,'DepDel15', 'CRSDepTime',
'DepDelayMinutes' ,'OriginAirportID' ,'DestAirportID' ,'ArrTime'
,'CRSArrTime' ,'ArrDel15' , 'ArrDelayMinutes','Origin','Dest']]
    df02=df01.loc[df01['Origin'].isin([
        'ATL',
        'CLT',
        'DEN',
        'DFW',
        'EWR',
        'IAH',
        'JFK',
        'LAS',
        'LAX',
        'MCO',
        'MIA',
        'ORD',
        'PHX',
        'SEA',
        'SFO'
        ])]
    df02=df02.loc[df02['Dest'].isin(['ATL','CLT',
        'DEN',
        'DFW',
        'EWR',
        'IAH',
        'JFK',
        'LAS',
        'LAX',
        'MCO',
        'MIA',
        'ORD',
        'PHX',
        'SEA',
        'SFO'])]
      
    see1=pd.concat([df02,see1])
```

```python
see2=pd.DataFrame()

for i in range(0,len(csv_files2)):
  
    df1 = pd.read_csv(csv_files2[i])
    df11=df1[['FlightDate', 'Quarter' ,'Year','Month',
  'DayofMonth' ,'DepTime' ,'DepDel15', 'CRSDepTime',
  'DepDelayMinutes' ,'OriginAirportID' ,'DestAirportID' ,'ArrTime'
  ,'CRSArrTime' ,'ArrDel15' , 'ArrDelayMinutes','Origin','Dest']]
    
    df12=df11.loc[df11['Origin'].isin([
          'ATL',
          'CLT',
          'DEN',
          'DFW',
          'EWR',
          'IAH',
          'JFK',
          'LAS',
          'LAX',
          'MCO',
          'MIA',
          'ORD',
          'PHX',
          'SEA',
          'SFO'
          ])]
  
    df12=df12.loc[df12['Dest'].isin(['ATL','CLT',
          'DEN',
          'DFW',
          'EWR',
          'IAH',
          'JFK',
          'LAS',
          'LAX',
          'MCO',
          'MIA',
          'ORD',
          'PHX',
          'SEA',
          'SFO'])]
        
    see2=pd.concat([df12,see2])
```

```python
final=pd.concat([see1,see2])
final.reset_index(drop=True, inplace=True)
```

```python
dump=[]
for i in final['CRSArrTime']:
  rr=round(i/ 100) * 100
  dump.append(rr)
```
